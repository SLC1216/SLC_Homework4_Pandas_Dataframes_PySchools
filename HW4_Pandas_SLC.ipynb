```python
# Using loc method to determine the number who passed in math
df_passedrows=df_studentsinschools.loc[df_studentsinschools['math_score'] > 69]
df_passedrows
```

```python
#Set variable to the total passed
totalpassed=df_passedrows.count()
totalpassed[10]
```

```python
#Divide total passed, variable "totalpassed" by the total student, variable "studentcount" and multiply by 100 to obtain percent passed
percentpassmath=totalpassed/studentcount*100
percentpassmath[10]
```

```python
# Using loc method to determine the number who passed in reading
df_passedrows_reading=df_studentsinschools.loc[df_studentsinschools['reading_score'] > 69]
df_passedrows_reading
```

```python
#Set variable to the total passed
totalpassed_reading=df_passedrows_reading['reading_score'].count()
totalpassed_reading
```

```python
#Divide total passed, variable "totalpassed" by the total student, variable "studentcount" and multiply by 100 to obtain percent passed
percentpassreading=totalpassed_reading/studentcount*100
percentpassreading[10]
```

```python
#Overall passing rate, avg of the two above percentages
overallpassrate=(percentpassmath+percentpassreading)/2
#(84.73+96.2)/2
overallpassrate[0]
```
